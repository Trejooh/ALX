```python
import pandas as pd 
import numpy as np
np.random.seed(0)
species_counts = np.random.randint(100, 300, size=100)
areas = ['Area' + str(i) for i in range(1, 101)]
data = {'Area': areas, 'Species_Count': species_counts}
df = pd.DataFrame(data)
population_mean = df['Species_Count'].mean()
sample = df.head()
sample_mean = sample['Species_Count'].mean()
print("Population Mean:", population_mean)
print("Sample Mean:", sample_mean)
```

```python
poulation_mean = df['Species_Count'].mean()
sample_mean = sample['Species_Count'].mean()
population_variance = df['Species_Count'].var(ddof=0)
sample_variance = sample['Species_Count'].var(ddof=1)
print("Population Mean:", population_mean)
print("Population Variance:", population_variance)
print("Sample Mean:", sample_mean)
```

```python
import matplotlib.pyplot as plt
sample_means = [np.mean(df['Species_Count'].sample(15)) for _ in range (100)]
plt.hist(sample_means, bins=10, edgecolor='black')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means')
plt.show()
```

Confidence Intervals(C)

```python
import scipy.stats as stats
sample_data = sample['Species_Count']
mean = np.mean(sample_data)
standard_error = stats.sem(sample_data)
degrees_freedom = len(sample_data) - 1
confidence_level = 0.95
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, mean, standard_error)
print(f"95% Confidence interval: {confidence_interval}")
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import pandas as pd
x = np.arange(1.2,2.2,0.01)
mu = 1.7
sigma = 0.1
f = st.norm.pdf(x, loc = mu, scale = sigma)
p = st.norm.cdf(1.5, loc = mu, scale = sigma)
print(f'The probability that a random data scientist has a height of 1.5 m is {np.round(p*100,2)} % ')
```

```python
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(x,f,'k' , label = 'Height normal distribution')
plt.axvline(x = 1.5, color = 'r', linestyle = '--', label = '1.5 m Tall')
plt.legend()
plt.show()
```
